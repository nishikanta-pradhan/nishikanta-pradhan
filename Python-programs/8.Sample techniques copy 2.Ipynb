### **`Sampling Techniques – Mobile App User Dataset`**
#### =========================================

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("/Users/rashmithav/Desktop/Desktop/datasets_dataminds/user_behavior_missing data-2.csv")
df
```

### 1️⃣ Probability Sampling

```python
# 1. Simple Random Sampling
simple_random = df.sample(n=10, random_state=42)
print("Simple Random Sample (User ID, Device Model, Gender):")
print(simple_random[['User ID','Device Model','Gender']], "\n")
```

```python
# 2. Stratified Sampling (by Gender)
df_gender = df.dropna(subset=['Gender'])
stratified = df_gender.groupby('Gender', group_keys=False)\
    .apply(lambda x: x.sample(frac=0.3, random_state=42))
print("Stratified Sample (Gender, App Usage Time, Battery Drain):")
print(stratified[['Gender','App Usage Time (min/day)','Battery Drain (mAh/day)']], "\n")
```

```python

# 3. Systematic Sampling (every 5th user)
step = 5
start = np.random.randint(0, step)
systematic = df.iloc[start::step]
print("Systematic Sample (User ID, App Usage Time, Screen On Time):")
print(systematic[['User ID','App Usage Time (min/day)','Screen On Time (hours/day)']], "\n")
```

```python
# 4. Cluster Sampling (select all users of one OS)
clusters = df['Operating System'].unique()
chosen_cluster = np.random.choice(clusters, size=1, replace=False)
cluster_sample = df[df['Operating System'].isin(chosen_cluster)]
print("Cluster Sample (Operating System, Device Model, App Usage Time):")
print(cluster_sample[['Operating System','Device Model','App Usage Time (min/day)']], "\n")
```

### 2️⃣ Non-Probability Sampling

```python
# 1. Accidental / Convenience Sampling (first 10 users)
accidental = df.head(10)
print("Accidental Sample (User ID, Device Model, Gender):")
print(accidental[['User ID','Device Model','Gender']], "\n")
```

```python
# 2. Judgmental / Purposive Sampling (high App Usage & Battery Drain)
judgmental = df[(df['App Usage Time (min/day)'] > 300) & (df['Battery Drain (mAh/day)'] > 1500)]
print("Judgmental Sample (App Usage Time, Battery Drain, Data Usage):")
print(judgmental[['App Usage Time (min/day)','Battery Drain (mAh/day)','Data Usage (MB/day)']], "\n")
```

```python
# 3. Quota Sampling (5 Male + 5 Female users)
quota = df_gender.groupby('Gender', group_keys=False)\
                 .apply(lambda x: x.sample(n=min(5,len(x)), random_state=42))
print("Quota Sample (Gender, Age, App Usage Time):")
print(quota[['Gender','Age','App Usage Time (min/day)']], "\n")
```

```python

# 4. Snowball Sampling (start with high App Usage, recruit high Screen On Time)
seed = df[df['App Usage Time (min/day)'] > 350].sample(1, random_state=42)
snowball_ids = set(seed['User ID'])
new_wave = df[df['Screen On Time (hours/day)'] > 6]
snowball_ids.update(new_wave['User ID'])
snowball = df[df['User ID'].isin(snowball_ids)]
print("Snowball Sample (App Usage Time, Screen On Time, Battery Drain):")
print(snowball[['App Usage Time (min/day)','Screen On Time (hours/day)','Battery Drain (mAh/day)']], "\n")
```

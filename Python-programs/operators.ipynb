```python
# arithmetic operation of two numbers
a = int(input("Enter number a : "))
b = int(input("Enter number b : "))
add = a + b
sub = a - b
mul = a * b
div = a / b
mod = a % b
quo = a // b
exp = a ** b
print("Arithmetic Opeations")
print("-------------------------")
print("Addition of",a ,"&", b," = ",add)
print("Substraction of",a ,"&", b," = ",sub)
print("Multiplication of",a ,"&", b," = ",mul)
print("Division of",a ,"&", b," = ",div)
print("Modulus of",a ,"&", b," = ",mod)
print("Quotient of",a ,"&", b," = ",quo)
print("Exponent of",a ,"&", b," = ",exp)
```

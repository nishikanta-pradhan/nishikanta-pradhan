```python
# 1️⃣ Single Inheritance
class Parent:
    def show(self):
        print("This is the Parent class.")

class Child(Parent):
    def display(self):
        print("This is the Child class.")

c = Child()
c.show()
c.display()
```

```python
# 2️⃣ Multiple Inheritance
class Father:
    def skills(self):
        print("Father: Knows gardening.")

class Mother:
    def hobbies(self):
        print("Mother: Loves painting.")

class Child(Father, Mother):
    def own_skill(self):
        print("Child: Plays guitar.")

c = Child()
c.skills()
c.hobbies()
c.own_skill()
```

```python
# 3️⃣ Multilevel Inheritance
class Grandparent:
    def legacy(self):
        print("Grandparent: Passed family values.")

class Parent(Grandparent):
    def guide(self):
        print("Parent: Teaches responsibility.")

class Child(Parent):
    def learn(self):
        print("Child: Learns everything.")

c = Child()
c.legacy()
c.guide()
c.learn()
```

```python
# 4️⃣ Hierarchical Inheritance
class Parent:
    def message(self):
        print("Parent: General message.")

class Son(Parent):
    def son_msg(self):
        print("Son: Specific message from Son.")

class Daughter(Parent):
    def daughter_msg(self):
        print("Daughter: Specific message from Daughter.")

s = Son()
d = Daughter()

s.message()
s.son_msg()

d.message()
d.daughter_msg()
```

# Abstraction

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    @staticmethod
    def hello():  
        print("Welcome to DataMinds clg") 

    def cal_average(self):
        total = sum(self.marks)
        print("Hi", self.name, "your avg score is", total / len(self.marks))

st1 = Student("Alice", [85, 90, 95])
st1.hello()       
st1.cal_average()
```

# Polymorphism

```python
class Student:
    def study(self):
        print("The student is studying...")

class OnlineStudent(Student):
    def study(self):
        print("The student studies through online classes.")

class OfflineStudent(Student):
    def study(self):
        print("The student studies in the classroom.")

# Function using polymorphism
def show_study_style(student):
    student.study()

# ✅ Run-time polymorphism in action
s1 = OnlineStudent()
s2 = OfflineStudent()

show_study_style(s1)  
show_study_style(s2)
```

# Encapsulation

```python


class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner 
        self.__balance = balance  

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount

    def get_balance(self):
        return self.__balance

account = BankAccount("John Doe", 1000)
print(account.owner)  

account.deposit(500)
account.withdraw(200)
account.withdraw(15000)
print(account.get_balance()) 
```

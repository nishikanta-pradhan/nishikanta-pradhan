#### **Real time example:**
## **Where we need to analyze sales data for different products across several months.**


```python
# Step 1: Sales data for 3 products across 6 months
import numpy as np
sales_data = {
    'Product_A': np.array([250, 300, 400, 500, 600, 700]),
    'Product_B': np.array([150, 200, 250, 300, 350, 400]),
    'Product_C': np.array([100, 150, 200, 250, 300, 350])
}
```

```python
# Step 2: Calculate total sales for each product
total_sales = {product: np.sum(sales) for product, sales in sales_data.items()}
total_sales
```

```python
# Step 3: Calculate average monthly sales for each product
average_sales = {product: np.mean(sales) for product, sales in sales_data.items()}
average_sales
```

```python
# Step 4: Find the product with highest sales in 4th month (April-like case)
month_index = 3  # 0-based index (0 = Jan, 1 = Feb, 2 = Mar, 3 = Apr)
sales_in_fourth_month = {product: sales[month_index] for product, sales in sales_data.items()}
highest_sales_product = max(sales_in_fourth_month, key=sales_in_fourth_month.get)
sales_in_fourth_month 
highest_sales_product 
```

```python
# Step 5: Print the output in readable format
print("📊 Total Sales per Product:")
for product, total in total_sales.items():
    print(f"{product}: {total} units")

print("\n📈 Average Monthly Sales:")
for product, avg in average_sales.items():
    print(f"{product}: {avg:.2f} units")

print("\n🏆 Highest Sales in 4th Month:")
print(f"{highest_sales_product} with {sales_in_fourth_month[highest_sales_product]} units sold")
```

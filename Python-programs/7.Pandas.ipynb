```python
# 2️⃣ Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Evan'],
    'Age': [25, 30, 35, np.nan, 28],
    'Salary': [50000, 60000, 70000, 80000, np.nan],
    'Department': ['HR', 'Finance', 'IT', 'IT', 'Finance'],
    'Joining_Date': ['2020-01-15', '2018-07-10', '2019-03-22', '2021-11-01', '2022-06-30']
}

df = pd.DataFrame(data)
df
```

```python
df['Joining_Date'] = pd.to_datetime(df['Joining_Date'])

print(" DataFrame Created:")
df['Joining_Date'] 
```

```python
# 5️⃣ Create new columns

today = pd.Timestamp('2025-11-04')   
df['Experience'] = (today - df['Joining_Date']).dt.days / 365
df['Experience']
# ➡️ Subtracts each employee’s Joining_Date from today. This gives a time difference (a pandas Timedelta object).
```

```python
# Add a new column: Salary category
df['Salary_Level'] = pd.cut(df['Salary'], bins=3, labels=['Low', 'Medium', 'High'])
print(df[['Name', 'Salary', 'Experience', 'Salary_Level']])
```

- **`GrouoBy`**


```python
# Let’s start simple — find the average salary per department
avg_salary = df.groupby('Department')['Salary'].mean()
print("\nAverage Salary per Department:")
avg_salary
```

- **`Multiple Aggregations`**
- You can also apply multiple functions at once using .agg() — for example, average salary and total employees per department.

```python
dept_summary = df.groupby('Department').agg({
    'Salary': ['mean', 'sum', 'max'],
    'Experience': ['mean', 'count']
})
print("\nDepartment Summary (Multiple Aggregations):")
dept_summary
```

- **`Group By Multiple Columns`**


```python
salary_by_dept_exp = df.groupby(['Department', 'Experience'])['Salary'].sum()
print("\nSalary Grouped by Department and Experience:")
salary_by_dept_exp
```

- **`Using groupby() with Filters`**
    

```python
high_salary = df.groupby('Department')['Salary'].mean().reset_index()
high_salary = high_salary[high_salary['Salary'] > 60000]
print("\nDepartments with Average Salary > 60,000:")
print(high_salary)
```

#### **What is Regular Expression (RegEx)?**


```python
# You can use RegEx through the re module.
import re

text = "My email is data.minds@gmail.com"

# Find email pattern
pattern = r"\S+@\S+\.\S+"  # Rule for email
result = re.findall(pattern, text)
print(result)

# “Here, \S+ means one or more characters that aren’t spaces,
# and @ + . make sure it matches an email format!”
```

```python
import re
# Find all digits in text
text = "There are 3 cats and 12 dogs."
result = re.findall(r'\d+', text)
print(result)
```

```python
text = "cat, car, dog, cup, bat"
result = re.findall(r'c\w+', text)
print(result)
# r'c\w+' → finds all words that start with the letter ‘c’ followed by letters or numbers.
```

#### **`Common RegEx Patterns`**

| Pattern | Meaning                          | Example Match                   |
| ------- | -------------------------------- | ------------------------------- |
| `\d`    | Digit (0–9)                      | “123”                           |
| `\D`    | Non-digit                        | “abc”                           |
| `\w`    | Word character (letters/numbers) | “Hello123”                      |
| `\s`    | Space                            | “ ”                             |
| `.`     | Any character                    | “a”, “1”, “#”                   |
| `^`     | Starts with                      | `^Hello` → “Hello World”        |
| `$`     | Ends with                        | `world$` → “Hi world”           |
| `+`     | One or more                      | `\d+` → “12345”                 |
| `*`     | Zero or more                     | `go*` → “g”, “go”, “gooo”       |
| `?`     | Optional                         | `colou?r` → “color” or “colour” |



🧹 Real-Life Uses in Data Science

    ✅ Cleaning messy text (emails, phone numbers, IDs)
    ✅ Extracting patterns (like dates or hashtags)
    ✅ Validating input (like postal codes, PAN numbers)
    ✅ Tokenizing text for NLP


```python
import re

text = "My phone numbers are 9876543210 and 9123456789."

pattern = r"\d{10}"  # 10 digits in a row
numbers = re.findall(pattern, text)
print(numbers)
# See that? It picked only the 10-digit numbers! That’s the power of patterns!”
```

### **Let me to show how RegEx is used inside pandas (for example — cleaning a DataFrame column using str.contains() or str.replace())**

```python
import pandas as pd

data = {
    'Email': ['data.minds@gmail.com', 'hello_world@yahoo.com', 'contact@dataminds.in', 'test123@outlook.com'],
    'Message': ['Order ID: 12345', 'Refund ID: 54321', 'Order ID: 67890', 'Promo code: SAVE50']
}

df = pd.DataFrame(data)
print("Original Data:")
df
```

- **`1: Using RegEx with str.contains()`**
- 👉 Goal: Find all rows where the message contains the word “Order”

```python
order_rows = df[df['Message'].str.contains(r'Order', regex=True)]
print(order_rows)
```

- **`2: Using RegEx with str.replace()`**
- 👉 Goal: Extract only the numeric Order IDs from the Message column

```python
df['Order_Number'] = df['Message'].str.replace(r'\D', '', regex=True)
print(df)
# “\D means non-digit. So str.replace(r'\D', '', regex=True) removes everything except numbers.”
```

- **`3: Extracting Patterns with str.extract()`**
- 👉 Goal: Pull out only the email domain names (like gmail.com, yahoo.com)

```python
df['Domain'] = df['Email'].str.extract(r'@([\w\.]+)')
print(df[['Email', 'Domain']])

# “Here, @([\w\.]+) means:
# Find @,
# Then capture all letters, digits, underscores, and dots after it.”
```

| Function         | Purpose                             | Example                          |
| ---------------- | ----------------------------------- | -------------------------------- |
| `str.contains()` | Filter rows by pattern              | Find emails containing ‘gmail’   |
| `str.replace()`  | Clean or remove unwanted characters | Remove special symbols           |
| `str.extract()`  | Capture specific pattern            | Extract domain, ID, or code      |
| `str.match()`    | Match entire string pattern         | Validate phone numbers           |
| `str.split()`    | Split text by pattern               | Split full names into first/last |

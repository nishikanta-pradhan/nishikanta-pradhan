```python
import random
secret_number = random.randint(1, 10)
guess = None

while guess != secret_number:
    guess = int(input("Guess the number between 1 and 10: "))
    
    if guess < secret_number:
        print("Too low!")
    elif guess > secret_number:
        print("Too high!")

print("Congratulations! You guessed it.")
```

```python
# List of available pizzas
pizzas = ["Margherita", "Pepperoni", "Veggie Delight", "BBQ Chicken", "Paneer Tikka"]

print("🍕 Welcome to Python Pizza Shop! Here’s our menu:")
for pizza in pizzas:
    print(f"- {pizza}")

# Step 2: Take Orders using While Loop
orders = []
ordering = True

while ordering:
    choice = input("\nWhich pizza would you like to order? (Type 'stop' to finish): ")
    
    if choice.lower() == "stop":
        ordering = False
    elif choice in pizzas:
        orders.append(choice)
        print(f"✅ {choice} added to your order!")
    else:
        print("❌ Sorry, we don't have that pizza. Please choose from the menu.")

# Step 3: Show Final Order
print("\n🧾 Here’s your final order summary:")
for order in orders:
    print(f"- {order}")

print("\n🎉 Thank you for ordering with Python Pizza Shop!")
```

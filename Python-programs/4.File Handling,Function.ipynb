```python
# Reading from the file to verify the contents
with open('file2.txt', 'r') as file:
    content = file.read()
    print("File Content After Append:\n", content)
```

```python
# Writing to the file normally (overwrites existing content)
with open('file2.txt', 'w') as file:
    file.write("This will overwrite the file content.\n")
```

```python
# Reading the file after overwrite
with open('file2.txt', 'r') as file:
    content = file.read()
    print("File Content After Overwrite:\n", content)
```

```python
# Reading and writing to the file without overwriting (r+ mode)
with open('file2.txt', 'r+') as file:
    file.write("Modifying the file while reading.\n")
    file.seek(5)  # Move the pointer back to the start of the file to read from the beginning
    content = file.read()
    print("File Content After r+ Mode:\n", content)
```

```python
# Reading the file after overwrite
with open('file2.txt', 'r') as file:
    content = file.read()
    print("File Content After Overwrite:\n", content)
```

```python
# Appending and reading at the same time (a+ mode)
with open('file2.txt', 'a+') as file:
    file.write("Appending again and reading together.\n")
    file.seek(0)  # Move pointer to start for reading
    content = file.read()
    print("File Content After a+ Mode:\n", content)
```

```python
# Open the file in read and write mode ('r+')
with open('file2.txt', 'r+') as file:
    print("Initial cursor position:", file.tell())  
    write_length = file.write("Hello Everyone ")  
    print("Number of characters written:", write_length)

    file.write("This is Python class ") 
    print("Cursor position after writing:", file.tell())  

    file.seek(0)  
    content = file.read()  

    print("File content:") 
    print(content)          
      
    file.close()  
```

```python
from PIL import Image
import os

# Original image path
original_image_path = '/Users/rashmithav/Desktop/File handling.jpeg'

# New folder path where you want to save the image
new_folder_path = '/Users/rashmithav/Desktop/Handling'

# Create the folder if it doesn't exist
os.makedirs(new_folder_path, exist_ok=True)

# Open the original image
img = Image.open(original_image_path)

# Define the new image path inside the new folder
new_image_path = os.path.join(new_folder_path, 'Handling.jpeg')

# Save the image to the new folder
img.save(new_image_path)

print(f"Image successfully saved to: {new_image_path}")
```

```python
# Traditional way
numbers = [1, 2, 3, 4, 5]
squares = []
for n in numbers:
    squares.append(n**2)
print(squares)   # [1, 4, 9, 16, 25]

```

```python
# List Comprehension way
squares = [n**2 for n in numbers]
print(squares)   # [1, 4, 9, 16, 25]
```

```python
def count_up_to(n):
    for i in range(1, n+1):
        yield i

for num in count_up_to(3):
    print(num)   
```

```python


# Multiple yield Statements
def my_generator():
    yield "Apple"
    yield "Banana"
    yield "Cherry"
    yield "Mango"

# Using the generator
for fruit in my_generator():
    print(fruit)
```
